```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
all_titles   = []
all_ranks    = []
all_websites = []
```

```python
#First website (bfi) All done!!!
url = 'https://www.bfi.org.uk/sight-and-sound/greatest-films-all-time'
r   = requests.get(url)
print("Status: " + str(r.status_code))
html_data = r.text
soup = BeautifulSoup(html_data, 'html5lib')

titles = soup.find_all('h1')[1:]

bfi_lst   = []
for i in range(len(titles)):
    bfi_lst.append(titles[i].text)

bfi_rank = list(range(1, len(bfi_lst) + 1))
bfi_rank.reverse()

all_titles.extend(bfi_lst)
all_ranks.extend(bfi_rank)
all_websites.extend(["Bfi"]*len(bfi_lst))
```

```python
#2nd Website (Collider) All Done!!!!
url = 'https://collider.com/top-highest-rated-movies-on-rotten-tomatoes/'
r   = requests.get(url)
print("Status: " + str(r.status_code))
html_data = r.text
soup = BeautifulSoup(html_data, 'html5lib')

titles = soup.find_all('h2')[:-1]

collider_lst   = []
collider_rank  = []

for i in range(len(titles)):
    collider_lst.append( titles[i].text[:-8].split('. ',1)[1])
    collider_rank.append(int(titles[i].text.split('.',1)[0]))
    
all_titles.extend(collider_lst)
all_ranks.extend(collider_rank)
all_websites.extend(["Collider"]*len(collider_lst))
```

```python
def SearchMovie(movie_name, api_key):
    params = {"query": movie_name, "api_key": api_key}
    url = "https://api.themoviedb.org/3/search/movie"

    response = requests.get(url, params = params)
    print("status:", response.status_code)
    
    #Checks if the request status is "OK" and that there are at least one movies result returned
    if response.status_code == 200 and (len(response.json()['results']) > 0):
        data = response.json()
        return data['results'][0]
    else:
        return "Error"
```
